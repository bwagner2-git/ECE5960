```python
def testDataRate(amount):
    x=""
    for i in range(amount):
        x=x+str(i)
    # print(x)
    b=bytes(x,'utf-8')
    # print(type(b))
    numBytes=len(b)
    print('number of bytes sent ' + str(2*numBytes))
    start=time.time()
    ble.send_command(CMD.ECHO, x)
    end=time.time()
    dataRate=numBytes*2/(end-start) ###bytes are sent back so total number of bytes transmitted is 2*numBytes
    print('the data rate was '+ str(dataRate))
    return (numBytes*2,(end-start))
```

```python
import matplotlib.pyplot as plt
runs=[]
runs.append(testDataRate(10))
time.sleep(1)
runs.append(testDataRate(30))
time.sleep(1)
runs.append(testDataRate(75))

###plotting example found here https://stackoverflow.com/questions/18458734/how-do-i-plot-list-of-tuples-in-python
x_val = [x[0] for x in runs]
y_val = [x[1] for x in runs]

plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.show()

```
